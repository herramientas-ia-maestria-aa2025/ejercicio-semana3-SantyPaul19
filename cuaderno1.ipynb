```python
from faker import Faker
fake = Faker()
```

```python
lista = []
```

```python
data = pd.DataFrame(lista)
```

```python
import random
```

```python
def generar_datos_enteros(minimo, maximo, cantidad):
  """Genera una lista de 'cantidad' números enteros aleatorios entre 'minimo' y 'maximo'."""
  datos = []
  for _ in range(cantidad):
    datos.append(random.randint(minimo, maximo))
  return datos
```

```python
lista2 = generar_datos_enteros(30, 50, 1000)
```

```python
data['edad'] = lista2
```

```python
data
```

```python
def crear_lista_random_de_valores(valores):
    
    return [random.choice(valores) for _ in range(1000)]
```

```python
ciudades = ["Pichincha", "Guayas", "Azuay", "Loja", "Manabi"]
ciudades_random = crear_lista_random_de_valores(ciudades)
```

```python
data['ciudades'] = ciudades_random
```
